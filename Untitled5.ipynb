```python
from datetime import datetime
from hijri_converter import convert

# Function to convert Gregorian date to Hijri date in YYYYMMDD format
def convert_to_hijri(gregorian_date):
    date_obj = datetime.strptime(gregorian_date, '%Y-%m-%d')
    hijri_date = convert.Gregorian(date_obj.year, date_obj.month, date_obj.day).to_hijri()
    return f"{hijri_date.year:04d}{hijri_date.month:02d}{hijri_date.day:02d}"

# Example usage
gregorian_dates = [
      "2025-10-18"
   ]
hijri_dates = [convert_to_hijri(date) for date in gregorian_dates]
for g_date, h_date in zip(gregorian_dates, hijri_dates):
    print(f"Gregorian Date: {g_date} -> Hijri Date: {h_date}\n")
```
